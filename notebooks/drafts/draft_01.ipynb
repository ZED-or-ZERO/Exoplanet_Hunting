```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

flux_df = pd.read_csv('../../data/processed/clean_flux.csv')

planet_idx = flux_df[flux_df['class'] == 1].index[0]
star_flux = flux_df.drop(columns=['id', 'class']).iloc[planet_idx].values

flux_image = np.array([star_flux])

plt.figure(figsize=(15, 2)) 
plt.imshow(flux_image, aspect='auto', cmap='afmhot')

plt.title('1D Flux Heatmap (Stellar Brightness Barcode)', color='white')
plt.xlabel('Time Steps')
plt.yticks([])
plt.colorbar(label='Normalized Brightness')

plt.gca().set_facecolor('black')
plt.gcf().patch.set_facecolor('black')
plt.tick_params(colors='white')
plt.xlabel('Time Steps', color='white')

plt.show()
```
