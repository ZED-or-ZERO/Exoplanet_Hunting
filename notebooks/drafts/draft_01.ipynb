```python
import pandas as pd

# Загружаем полный очищенный датасет
flux_df = pd.read_csv('../../data/processed/clean_flux.csv')

# Берем одну звезду, у которой точно есть экзопланета (class 1)
star_with_planet = flux_df[flux_df['class'] == 1].drop(columns=['id', 'class']).iloc[[0]]
star_with_planet.to_csv('../../data/processed/demo_star_planet.csv', index=False)

# Берем одну звезду, у которой точно НЕТ экзопланеты (class 0)
star_no_planet = flux_df[flux_df['class'] == 0].drop(columns=['id', 'class']).iloc[[0]]
star_no_planet.to_csv('../../data/processed/demo_star_noise.csv', index=False)

print("Demo files created successfully.")
```
